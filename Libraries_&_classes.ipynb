```python
class car:
  def __init__(self,brand,model):
    self.brand=brand
    self.model=model

car1=car("ford","mustang")
print(car1.brand)
print(car1.model)

car2=car("BMW","M5")
print(car2.brand)
print(car2.model)

```

```python
class student:
  def __init__(self,name, marks):
    self.name=name
    self.marks=marks
  def display(self):
    print(f"{self.name} scored {self.marks}")

student1=student("John",90)
student1.display()
```
